```python
#: Visualizing the Hilbert Matrix with a Heatmap
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Function to generate Hilbert matrix of size n x n
def hilbert_matrix(n):
    return np.array([[1 / (i + j - 1) for j in range(1, n + 1)] for i in range(1, n + 1)])

# Generate a 10x10 Hilbert matrix
H = hilbert_matrix(10)

# Create a heatmap to visualize the Hilbert matrix
sns.heatmap(H, annot=True, cmap="coolwarm", linewidths=0.5)

# Add title
plt.title("Hilbert Matrix Heatmap (10x10)")

# Show the plot
plt.show()
```
