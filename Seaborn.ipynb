```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

n = 200
D = np.zeros((n, 2))

D[:, 0] = np.linspace(0, 10, n) + np.random.randn(n)
D[:, 1] = D[:, 0]**2 + np.random.randn(n) * 10

# Correct usage for jointplot
sns.jointplot(x=D[:, 0], y=D[:, 1])
plt.show()
```

```python
df = pd.DataFrame(data=D,columns=['x','y'])

sns.jointplot(df.columns[0],df.columns[1])
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create custom data
np.random.seed(42)  # For reproducibility

# Generate random data
data = {
    "x": np.random.rand(50) * 100,  # Random values for x-axis (e.g., product prices)
    "y": np.random.rand(50) * 200,  # Random values for y-axis (e.g., sales numbers)
    "category": np.random.choice(['A', 'B', 'C'], size=50)  # Random categories
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Create a scatter plot using Seaborn
sns.scatterplot(x="x", y="y", hue="category", data=df)

# Show the plot
plt.title("Custom Data Scatter Plot")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create custom data
np.random.seed(42)  # For reproducibility

# Generate random data: 1000 data points from a normal distribution
data = np.random.randn(1000)

# Create a DataFrame (optional, for better visualization)
df = pd.DataFrame(data, columns=["values"])

# Create a KDE plot using Seaborn
sns.kdeplot(df["values"], shade=True, color="blue")

# Add title
plt.title("Kernel Density Estimate (KDE) Plot")

# Show the plot
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create custom data
np.random.seed(42)  # For reproducibility

# Generate random data: 1000 data points from a normal distribution
data = np.random.randn(1000)

# Create a DataFrame (optional, for better visualization)
df = pd.DataFrame(data, columns=["values"])

# Create a KDE plot using Seaborn with rainbow color map
sns.kdeplot(df["values"], fill=True, cmap="rainbow")

# Add title
plt.title("Kernel Density Estimate (KDE) Plot with Rainbow Color Map")

# Show the plot
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create custom data (matrix D)
np.random.seed(42)  # For reproducibility

# Create a 2D matrix D with random data (50 rows and 2 columns)
D = np.random.randn(50, 2) * 10  # Random data scaled by 10

# Convert the matrix D to a DataFrame df
df = pd.DataFrame(D, columns=["X", "Y"])

# Create a regression plot using Seaborn
sns.regplot(x="X", y="Y", data=df)

# Add title
plt.title("Regression Plot of Data in Matrix D")

# Show the plot
plt.show()
```
